# Problem 1

###Total grains received

```python
import matplotlib.pyplot as plt

def predictRiceGrains(days):
  today_grains = []
  list_of_days = []
  grains = 0
  for x in range(0, days):
    double = 2**x
    today_grains.append(double)
    grains = grains + double
    list_of_days.append(x+1)

  return [grains, today_grains, list_of_days]

total_grains, grains_per_day, list_of_days = predictRiceGrains(19)
print(f"In the 100th day he received in total: {total_grains} grains")
```

### Generate graph of Grains per day vs number of days

```python
def generateGraph(days, grainsPerDay):
  fig, axe = plt.subplots()
  axe.set_xlabel('Number of days')
  axe.set_ylabel('Grains per day')
  axe.plot(days, grainsPerDay,marker = 'o')
  plt.show()
generateGraph(list_of_days, grains_per_day)
```

This is an example proving that it is actually an line graphic but the number of elements present in the graph give it the shape of an exponential one

# Problem 2

```python
def compute_sorori_shinzaemon(n_days=100):
    list_n_grains = []
    list_total_grains = []
    list_total_days =[]
    for x in range (0, n_days):
      double = 2**x
      list_n_grains.append(double)
      total_grains = sum(list_n_grains)
      list_total_days.append(x)
      list_total_grains.append(total_grains)

    return list_n_grains, list_total_grains, list_total_days

# compute_sorori_shinzaemon(n_days=100)
list_n_grains, list_total_grains, list_total_days = compute_sorori_shinzaemon(n_days=19)
print(f"By attending to {19} days I would receive {list_total_grains[len(list_total_grains) - 1]}")

print(list_n_grains, list_total_grains)
```

```python
import matplotlib.pyplot as plt

fig, axe = plt.subplots()
axe.set_xlabel('Number of days:')
axe.set_ylabel('Grains per day')
axe.plot(list_total_days, list_n_grains,marker = 'o')
plt.show()
```

#Problem 3

##Research basis:
    - FAO/WHO recommended daily rice intake: 400g per person
    - Average weight of 1 rice grain: 0.029g
    - Daily grains needed per person: ~13,793

```python
def calculate_survival_days(rice_grains: int, num_people: int) -> int:
    GRAINS_PER_DAY_PER_PERSON = 13793

    if num_people <= 0:
        raise ValueError("Number of people must be positive")

    total_grains_needed_per_day = GRAINS_PER_DAY_PER_PERSON * num_people
    survival_days = rice_grains // total_grains_needed_per_day

    return survival_days

days = calculate_survival_days(1000000, 2)
print(f"Can survive for {days} days")
```
